```python
import pandas as pd
df = pd.read_csv("../../../3. Data/1. Raw_Features/features_only.csv")

df["pdLabel"] = (
    (df["bounceCount"] >= 3) |
    (df["emiRatio"] >= 0.50) |
    (df["expenseRatio"] >= 0.85)
).astype(int)

df.to_csv("../../../3. Data/2. PD_Data/features_with_pdlabel.csv", index=False)

print("pdLabel generated")
print(df["pdLabel"].value_counts(normalize=True))
```
